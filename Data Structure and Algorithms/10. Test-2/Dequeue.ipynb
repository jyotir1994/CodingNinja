```python


class Stack:
    
	# Initialize your data structure.
    def __init__(self):
        self.dq = []
        self.size=0
        self.limit=10
        

    # Pushes 'X' into the stack. Returns true if it gets pushed into the stack, and false otherwise.
    def isEmpty(self):
        return self.size == 0

    def push_front(self, x):
        if self.size==self.limit:
            print(-1)
            return
        if self.size==0:
            self.dq.append(x)
        else:
            self.dq.insert(0,x)
        self.size+=1    
        pass
    def push_rear(self,x):
        if self.size==self.limit:
            print(-1)
            return
        self.dq.append(x)
        self.size+=1
        pass
    

    # Pops top element from Stack. Returns -1 if the stack is empty, otherwise returns the popped element.
    def pop_front(self):
        if (len(self.dq) == 0):
            # Stack is empty.
            print(-1)
            return
        self.dq.pop(0)
        self.size-=1
        
    def pop_rear(self):
        if self.size==0:
            print(-1)
            return
        self.dq.pop()
        self.size-=1
        pass

    # Returns the topmost element of the stack. In case the stack is empty, it returns -1.
    def get_front(self):
        if self.size==0:
            return -1
        return self.dq[0]
    def get_rear(self):
        if self.size==0:
            return -1
        return self.dq[-1]
        

    # Returns true if the stack is empty, otherwise false.


    # Returns the number of elements currently present in the stack.
    def Size(self):
        return self.size
 
#main
DQ=Stack()
li=[int(x) for x in input().split()]
index=0
while len(li)!=0 and li[index]!=-1:
    char=li[index]
    if char==1:
        index+=1
        DQ.push_front(li[index])
    elif char==2:
        index+=1
        DQ.push_rear(li[index])
    elif char==3:
        DQ.pop_front()
        pass
    elif char==4:
        DQ.pop_rear()
        pass
    elif char==5:
        print(DQ.get_front())
        
        pass
    elif char==6:
        print(DQ.get_rear())
        pass
    index+=1
```
